```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
import re
```

```python
#yearList = ["2011", "2012", "2013", "2014"]
yearList = ["2015", "2016","2017","2018"]
weatherList = []
```

```python
weatherTranslate = {
                        '晴': "clear",
                        '多云': "partly cloudy",
                        '雨夹雪': "sleet",
                        '阴': "cloudy",
                        '小雨': "light rain",
                        '小到中雪': "small to medium snow",
                        '小雪': "light snow",
                        '大雪': "heavy snow",
                        '中到大雪': "medium to heavy snow",
                        '中雨': "rain",
                        '阵雨': "shower",
                        '雷阵雨': "thunder storm",
                        '中到大雨': "medium to heavy rain",
                        '大雨': "heavy rain",
                        '大到暴雨': "heavy rain",
                        '小到中雨': "small to moderate rain",
                        '暴雨': "rainstorm",
                        '-': "n/a",
                        '雾': "fog",
                        '': "n/a"
                    }
```

```python
def unique(list1): 
  
    unique_list = [] 
      
    for x in list1: 
        # check if exists in unique_list or not 
        if x not in unique_list: 
            unique_list.append(x) 
    return unique_list
```

```python

for year in yearList:
    
    rows = []
    
    # all months
    for i in range(1, 13):

        if i < 10:
            m = "0" + str(i)
        else:
            m = "" + str(i)
    
        quote_page = 'http://www.tianqihoubao.com/lishi/shanghai/month/'+year+m+'.html'
        print(quote_page)
        page = urlopen(quote_page)


        soup = BeautifulSoup(page, 'html.parser', fromEncoding="gb2312")

        for row in soup.find_all('tr'):
            columns = []
            
            colCount = 0
            for val in row.find_all('td'):
                text = val.text.replace(" ", "").replace("\r", "").replace("\n", "")
                
                if colCount == 0:
                    text = text.replace('年', '-')
                    text = text.replace('月', '-')
                    text = text.replace('日', '')
                
                elif colCount == 1:
                    weatherDesc = text.split("/")
                    newWeatherDesc = []
                    if len(weatherDesc) == 2:
                        weatherList.append(weatherDesc[0])
                        weatherList.append(weatherDesc[1])
                        
                        newWeatherDesc.append(weatherTranslate[weatherDesc[0]])
                        newWeatherDesc.append(weatherTranslate[weatherDesc[1]])
                        
                        text = ",".join(newWeatherDesc)
                
                elif colCount == 2:
                    minMaxTemperature = list(map(int, re.findall(r'\-*\d+', text)))
                    
                    if len(minMaxTemperature) == 2:
                        text = ",".join(str(v) for v in minMaxTemperature)
                    elif len(minMaxTemperature) == 1:
                        text = ",".join([str(minMaxTemperature[0]), str(minMaxTemperature[0])])
                    
                elif colCount == 3:
                    windSpeed = list(map(int, re.findall(r'\d+', text)))
                    
                    if len(windSpeed) == 0:
                        text = "0"
                    else:
                        text = str(max(windSpeed))
                        
                columns.append(text)
                

                colCount = colCount + 1
            
            # not include header
            if(len(columns) > 2):
                rows.append(columns)


    f = open("ScrapedWeather/"+year+".csv", 'w+')
    for row in rows:
        f.write("%s\n" % (",".join(row)))

    f.close()

```
